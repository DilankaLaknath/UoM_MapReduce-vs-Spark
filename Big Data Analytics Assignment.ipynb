```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = ["1", "2", "3", "4", "5"]
hadoop_y = [24.666, 12.889, 13.709, 13.073, 12.446]
spark_y = [15.30104327, 1.111656427, 1.065317631, 0.826749563, 0.91268754]

# Set the width of the bars
barWidth = 0.35

# Set the position of the bars on the x-axis
r1 = np.arange(len(hadoop_y))
r2 = [x + barWidth for x in r1]

# Create the double bar graph
plt.bar(r1, hadoop_y, color='red', width=barWidth, label='Hadoop')
plt.bar(r2, spark_y, color='blue', width=barWidth, label='Spark')

# Add labels and title
plt.xlabel("Number of iterations")
plt.ylabel("Running Time (s)")
plt.title("Comparison of Hadoop vs Spark for Carrier Delay")

# Set x-axis tick labels
plt.xticks([r + barWidth/2 for r in range(len(hadoop_y))], x)

# Add legend
plt.legend()

# Show the graph
plt.show()
```
